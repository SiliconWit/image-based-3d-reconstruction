```python
import numpy as np
import matplotlib.pyplot as plt
import os
# import OpenEXR
# import Imath
%matplotlib notebook
```

```python
# Enable OpenCV EXR support (must be set before importing cv2)
os.environ["OPENCV_IO_ENABLE_OPENEXR"] = "1"
import cv2
```

```python
# Function to read an EXR file using OpenCV
def read_exr_with_opencv(file_path):
    # Load the EXR file as a floating-point image
    exr_image = cv2.imread(file_path, cv2.IMREAD_ANYCOLOR | cv2.IMREAD_ANYDEPTH)
    
    if exr_image is None:
        raise ValueError(f"Failed to load EXR file: {file_path}")
    
    return exr_image

# Path to your EXR file
exr_file_path = "DepthCapture.exr"

try:
    # Read the EXR file
    depth_map = read_exr_with_opencv(exr_file_path)
    

    # Display the depth map using matplotlib
    plt.figure(figsize=(8, 6))
    plt.imshow(depth_map, cmap="viridis")
    plt.colorbar(label="Depth")
    plt.title("Depth Map")
    plt.axis("off")
    plt.show()

except Exception as e:
    print(e)
```
