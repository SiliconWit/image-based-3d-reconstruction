```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data from the Excel file
# Skip the first 2 rows since headers start at row 3 (indexing starts from 0)
data = pd.read_excel("PreliminaryExperimentResults.xlsx", skiprows=2)
```

```python
# Filter rows based on the conditions
filtered_data = data[
    (data["Model"] == "UtahTeapot") &
    (data["PATTERN"] == "Circular") &
    (data["LAMP"] == 0)
]
```

```python
filtered_data
```

```python
# Create a 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Parameters for the 3D bars
bar_width = 0.1
bar_depth = 0.1
z_base = 0

# Plotting 3D bars
ax.bar3d(filtered_data["RADIUS"], filtered_data["PHOTOS"], z_base, 
         bar_width, bar_depth, filtered_data["RMS"], shade=True)

# Setting the labels
ax.set_xlabel('RADIUS')
ax.set_ylabel('PHOTOS')
ax.set_zlabel('RMS')
ax.set_title('3D Bars of RADIUS, PHOTOS, and RMS')

plt.show()
```
