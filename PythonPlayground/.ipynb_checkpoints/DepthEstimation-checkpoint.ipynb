# Wrong Depth
Consider Test Image "singleQR.jpg". 
From the depth map generated by zoedepth, the QR poster at the bottom right is nearly 4 meters away. However from the ground truth the distance is about 4.5 meters away. 

Also, the closest point of the rabbit to the camera(face) is around 1.7meters away but from the ground truth, it is about 1.2 meters away. THis calls for the urgent need to look for alternative depth maps or find ways for calibratrion.

Below is a comparisong of ZoeDepth and GLPN

````python
# GLPN Inference function
from transformers import GLPNImageProcessor, GLPNForDepthEstimation
import torch
import numpy as np
from PIL import Image
import requests

def compute_depth_glpn(img_name):
    image = Image.open(f"testImages/{img_name}")
    processor = GLPNImageProcessor.from_pretrained("vinvino02/glpn-nyu")
    model = GLPNForDepthEstimation.from_pretrained("vinvino02/glpn-nyu")

    # prepare image for the model
    inputs = processor(images=image, return_tensors="pt")

    with torch.no_grad():
        outputs = model(**inputs)
        predicted_depth = outputs.predicted_depth

    # interpolate to original size
    prediction = torch.nn.functional.interpolate(
        predicted_depth.unsqueeze(1),
        size=image.size[::-1],
        mode="bicubic",
        align_corners=False,
    )

    return prediction.squeeze().cpu().numpy()
````

````python
img_name = "singleQR.jpg"
zoe_dep = compute_depth_standalone(load_image("singleQR.jpg"))
glpn_dep = compute_depth_glpn(img_name)
````

````python
test_points = list()
for i in range(5):
    x = np.random.randint(0, zoe_dep.shape[1] ) 
    y = np.random.randint(0, zoe_dep.shape[0] ) 
    test_points.append([x,y])

_, axs = plt.subplots( 1,2, figsize=(9,4), layout="constrained" )

for _img, ax, label in zip( [zoe_dep, glpn_dep], axs, ["ZoeDepth", "GLPN"] ):
    ax.imshow(_img)
    ax.set_title(label)
    ax.axis("off")
    for pt in test_points:
        dep_val = np.round(_img[pt[1],pt[0]],2)
        ax.scatter( pt[0], pt[1] , color='white', marker='o', s=10) 
        ax.text( pt[0], pt[1], str(dep_val), color='white', fontsize=10)
    ax.imshow(_img)
        
````

# Remeding the Challenges Above

Compared to Absolute Depth Estimators like ZoeDepth and GLPN a Relative Depth Estimators\
>"aim to regress pixel-wise depth predictions that are accurate relative to each other but carry no metric meaning."


## MiDAS v3.1

MiDAS is deep learning based Monocular Depth Estimator. It outputs a relative depth map given a simgle image. The latest version as of 12th Jan 2024 supports a number of pre-trained models with varying speeds and accuracies.


A scale factor and shift factor have to be determined first in order to obtain absolute values of the depth map.

````python
from midas.model_loader import load_model

def process(image, model, model_type, device, target_size, optimize):
    """
    Run the inference and interpolate.

    Args:
        device (torch.device): the torch device used
        model: the model used for inference
        model_type: the type of the model
        image: the image fed into the neural network
        input_size: the size (width, height) of the neural network input (for OpenVINO)
        target_size: the size (width, height) the neural network output is interpolated to
        optimize: optimize the model to half-floats on CUDA?

    Returns:
        the prediction
    """
    sample = torch.from_numpy(image).to(device).unsqueeze(0)

    if optimize and device == torch.device("cuda"):
        print("  Optimization to half-floats activated. Use with caution, because models like Swin require\n"
                    "  float precision to work properly and may yield non-finite depth values to some extent for\n"
                    "  half-floats.")
        sample = sample.to(memory_format=torch.channels_last)
        sample = sample.half()

    prediction = model.forward(sample)
    prediction = (
        torch.nn.functional.interpolate(
            prediction.unsqueeze(1),
            size=target_size[::-1],
            mode="bicubic",
            align_corners=False,
        )
        .squeeze()
        .cpu()
        .numpy()
    )

    return prediction

def compute_depth_relative(image):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
#     device = "cpu"
    print("Device: %s" % device)
#     model_type = "dpt_beit_large_384"
#     model_type = "dpt_swin2_large_384"
    model_type = "dpt_beit_large_512"
    model_path = f"Models/{model_type}.pt"
    model, transform, net_w, net_h = load_model(
        device, 
        model_path, 
        model_type, 
        optimize=True, 
        height=None, 
        square=False
    )

    original_image_rgb = image/255.0
    image = transform({"image": original_image_rgb})["image"]

    with torch.no_grad():
        prediction = process(image,
                             model, 
                             model_type, 
                             device,  
                             original_image_rgb.shape[1::-1],
                             True
                        )
    return prediction
````

# QR Code Markers and QReader

To obtain an absolute depth map from a relative one based on MiDaS,a shift and scale factor have to be determined. These can be determined given ground truth depth values from the same image. To this end, markers with known depth values can be incldued in the scene to serve as the ground truth points.

To this end, QR codes are one such choice of markers considered. The encoded content of the QR code chosen for this case is the position of the marker itself within the scene.

 
**QReader** is a python tool that leverages YOLOv8 and pyzbar to detect and work QR codes [[1]](https://pypi.org/project/qreader/).\
It does a better job than cv2's QRCodeDetector() which on the 3 qrcode image of the stanford bunny could only detect nothing. It is capable detecting QR codes in the scene and decoding them.

1. https://pypi.org/project/qreader/

````python
qr_img = load_image("vp5.jpg")
qreader = QReader()
res = qreader.detect_and_decode( image=qr_img, return_detections = True  )
detections = [index for index, element in enumerate(res[0]) if element != None]
````

# MultiQR with ZoeDepth

$ P_f = P *s + t $

````python
zoe_pts = list()
if 'zoe_dep' not in globals():
    zoe_dep = compute_depth_standalone(img)
    
for detection in detections:
    location = np.asarray([float(val) for val in res[0][detection].split(',')]) - cam_pose
    x, y = [ int(px_i) for px_i in res[1][detection]["cxcy"] ]
    zoe_pts.append({
        "depth"   : np.round(np.linalg.norm(location),6),
        "pred"    : float(zoe_dep[y,x]),
    })
    
# Using Least Squares
D = np.array([pt['depth'] for pt in zoe_pts])
P = np.array([pt['pred'] for pt in zoe_pts])

s, t = linear_regression_coefficients(D, P)
t_arr = np.full(img.shape[:2],t)
zoe_depc = zoe_dep * s + t_arr


_,axs = plt.subplots(1,3, layout="constrained", figsize=(9.7,3))

    
for _img, ax, label in zip( [abs_depth, zoe_dep, zoe_depc], axs, ["Depth","Zoe Depth", "ZD_Calib"] ):
    ax.imshow(_img)
    ax.set_title(label)
    ax.axis("off")
#     if label != "Real":
    for pt in test_points:
        dep_val = np.round(_img[pt[1],pt[0]],2)
        ax.scatter(pt[0], pt[1], color='white', marker='o', s=5)
        ax.text( pt[0], pt[1], str(dep_val), color='white', fontsize=10)
````
